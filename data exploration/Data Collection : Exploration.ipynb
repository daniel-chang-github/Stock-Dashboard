```python
##Deleted in the upload file.
API_KEY = ''
API_SECRET = ''
BEARER_TOKEN= ''

Access_Token= ''
Access_Token_Secret =''
```

```python
import tweepy

# Authenticate to Twitter
auth = tweepy.OAuthHandler(API_KEY, API_SECRET)
##auth.set_access_token("ACCESS_TOKEN", "ACCESS_TOKEN_SECRET")

# Create API object
api = tweepy.API(auth)
```

```python
##userid_track_list = 

#example https://twitter.com/emmetlsavage
userID = 'Benzinga'
```

```python
tweets = api.user_timeline(screen_name=userID, 
                           # 200 is the maximum allowed count
                           count=10,
                           include_rts = False,
                           # Necessary to keep full_text 
                           # otherwise only the first 140 words are extracted
                           tweet_mode = 'extended'
                          
                           )
## ,parser=tweepy.parsers.JSONParser()
```

```python
#View a sample tweet
tweets[0]._json 
```

```python
#Exporting a tweet sample to json
with open('tweepy.json', 'w') as outfile:
    json.dump(tweets[6]._json , outfile)
```
